_Примечание: Дата премьеры в формате ГГГГ-ММ-ДД_

```python
import bs4, pandas as pd, requests, re
response = requests.get('https://www.kino.ru/afisha/city/55441/page/1')
soup = bs4.BeautifulSoup(response.text)
articles = soup.find_all('article', class_='post')

def parse(article):
    title = article.find('h2').text
    title_orig = article.find('a', class_='h5').text
    info = article.find('div', class_='info-country').find_all('a')
    *genre, country, premiera_date = [i.text for i in info]
    rate = article.find('div', class_='rating').text
    
    return pd.Series({'title': title, 'title_orig': title_orig, 'country': country, 
                      'genre': genre, 'talks': '', 'rate': rate, 'premiera_date': premiera_date})


    
data = pd.DataFrame(list(map(parse, articles)))
data
```
